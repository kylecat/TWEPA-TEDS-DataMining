## 設定檔案路徑
* 檔案：TEDS10__POINT TWD97_Description.xls

```python
# 設定檔案路徑
dir_data = './TEDS_data/'
f_app = os.path.join(dir_data, 'TEDS10__Description.xls')
```

## 讀取檔案
* 使用pandas的read_excel() function來讀取資料
* read_excel()的參數說明：
    * 中文：https://codertw.com/%E7%A8%8B%E5%BC%8F%E8%AA%9E%E8%A8%80/568155/
    * 官網：https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html

```python
# sheetname=None 設定要開哪個分頁，None表示全部開啟(小心別爆掉了)
# sheetname為None或者是list時，read_excel()會返回成OrderedDict的資料型態
data = pd.read_excel(f_app, sheetname=None, index_col=0)
print("參數sheetname 設定為None或某個list時的資料型態:{}".format(type(data)))

data_1 = pd.read_excel(f_app, sheetname="POINT_CLASS", index_col=0)
print("參數sheetname 設定為第幾頁或某個分頁名稱時的資料型態:{}".format(type(data_1)))
```

## 讀取到的各分頁名稱

```python
# 資料型態OrderedDict必須用.items()才能把每一筆都抓出來
for k, v in data.items():
    print("分頁：{}\t資料型態：{}".format(k,type(v)))
```
